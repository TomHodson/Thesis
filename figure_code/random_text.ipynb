# Analogy to classical electromagnetism 

There are some nice analogies to be drawn with electromagnetism (or gauge field theories more generally if the reader is so inclined). Let's imagine we are doing classical magnetism in 2D with a B field $\vec{B} = (0,0,B_z(\vec{x}))$. The magnetic vector potential is then a divergence free quantity $\vec{A}$ defined such that $\nabla \times \vec{A} = \vec{B}$ or $B_z = \frac{\partial A_y}{\partial x} - \frac{\partial A_x}{\partial y}$. The flux through a region $S$ is related to an area integral of $B_z$ and a line integral of $\vec{A}$ via stokes' theorem:
$$
\int_S B_z \;dS = \int_{\partial S} A \cdot d\vec{l} 
$$

The Kitaev model is discrete so we don't have a direct analogue of the $B$ field but we can think of the fluxes through each plaquette as being related to an area integral of some imaginary smooth B field. 

We began in a spin model and we could define fluxes that are essentially a line integral around a loop, reminiscent of the fact that flux is the line integral of the magnetic vector potential

The Kitaev Honeycomb can be quite easily turned into a quantum error correcting code [like this](https://errorcorrectionzoo.org/c/honeycomb), the same idea applies to our model. 

In periodic boundaries there is an additional pair of global $\mathbb{Z}_2$ fluxes $\Phi_x$ and $\Phi_y$, which are calculated along an arbitrary closed path that wraps the torus in the $x$ and $y$ directions respectively. The energy difference between distinct flux sectors decays exponentially with system size, so that the ground state of any flux sector in the thermodynamic limit displays a __fourfold__ topological degeneracy [@kitaev_fault-tolerant_2003].
